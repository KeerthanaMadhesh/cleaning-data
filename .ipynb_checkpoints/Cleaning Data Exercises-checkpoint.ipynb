## Missing Data

In the Towards Data Science article, they list 3 techniques for detecting any missing data frame. Try all three before deciding on a way to solve any missing data issues.

```python
# Plot a heatmap of any missing data
plt.figure(figsize=(10,8))
cols = New.columns[:30]
colours = ['#000099', '#ffff00'] 
sns.heatmap(New[cols].isna(), cmap=sns.color_palette(colours))
```

```python
pct_missing = New.isna().mean()
pct_missing[:10]
```

```python
# Try out a missing data percentage list! 
pct_missing[pct_missing > .3]
```

```python
New_less_missing_cols = New.loc[:, pct_missing <= .3].copy() 
New_less_missing_cols.shape
```

```python
# Plot a missing data histogram
missing_by_row = New.isna().sum(axis='columns')
missing_by_row.hist(bins=50)
```

## Unnecessary Data

Unnecessary data could be duplicates, irrelevant, or any uninformative data. Dive into each type of unnecessary data to see if there is any in the data set. Make sure to make note of how you would handle each type of unnecessary data.

```python
# Look for any irrelevant data in the dataframe. How will you handle it?
New.drop(['Bridge Highway Name', 'Taxi Pick Up Location', 'Taxi Company Borough', 'Bridge Highway Direction', 'Road Ramp', 'Bridge Highway Segment', 'Garage Lot Name',	'Ferry Direction'], axis=1)

```

```python
# Think about what data in the dataframe may be uninformative. Make note here of what situations may render data uninformative and how you would handle it?
```

```python
we need to find the high percentage of same value which is uninformative and we need to drop them.
```

```python
num_rows = len(New)

for col in New.columns:
    cnts = New[col].value_counts(dropna=False)
    top_pct = (cnts/num_rows).iloc[0]
    
    if top_pct > 0.999:
        print('{0}: {1:.2f}%'.format(col, top_pct*100))
        print(cnts)
        print()
```

## Inconsistent Data

Inconsistent data is anything that messes with your model. This is likely due to inconsistent formatting and can be addressed by re-formatting all values in a column or row.

```python
# Try to convert all the strings in a column to lower or upper case. 
New['Agency'] = New['Agency'].str.lower()
New['Agency' ].value_counts(dropna=False)
```
